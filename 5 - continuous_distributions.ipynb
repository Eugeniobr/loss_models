```python
import numpy as np
from scipy.stats import poisson, gamma
rng = np.random.default_rng(42)
```

```python
a = 0.5
theta = 2.0
```

```python
lam = rng.gamma(shape=a, scale=theta, size=100000)
x = rng.poisson(lam=lam)
pmf_cond = poisson.pmf(x, mu=lam)     # P(X=x | 位=lam)
pdf_lambda = gamma.pdf(lam, a=a, scale=theta)  # f_位(lam)
```

```python
event_succ = [s for s in x if s == 1]
```

```python
print(len(event_succ)/len(x))
```
