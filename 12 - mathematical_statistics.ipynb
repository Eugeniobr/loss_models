```python
import numpy as np
import matplotlib.pyplot as plt

# parâmetros da U(-2, 2)
a, b = -2.0, 2.0
pdf_const = 0.25  # f(x) = 1 / (b - a)

# gerar amostras
n = 10000000
samples = np.random.uniform(a, b, size=n)

# exemplo de verificação visual
fig, ax = plt.subplots()
ax.hist(samples, bins=40, density=True, alpha=0.6, label="Amostras")
ax.hlines(pdf_const, a, b, colors="r", label="fdp teórica")
ax.set_xlim(a, b)
ax.legend()
plt.show()
```
